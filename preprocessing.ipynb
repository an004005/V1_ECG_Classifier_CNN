```python
import glob
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import operator
# import xml.etree.ElementTree as elemTree

#tensor modules
from sklearn.model_selection import train_test_split
```

```python
def cutUseless(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.fastNlMeansDenoising(img, None, 10, 7, 21)
    
    img = img[60:]
    img = img[:500]
    img = img.T
    img = img[123:]
    img = img[:660]
    img = img.T
    return img

def dimReduction(img):
    feature = []
    for i, row in enumerate(img.T):
        min_index, min_value = min(enumerate(row), key=operator.itemgetter(1))
        feature.append(min_index)
    if feature[-1] ==0:
        new_feature = []
        for i in range(330):
            new_feature.append(feature[i])
            new_feature.append(feature[i])
        return new_feature
    return feature
```

```python
Y = []
ecgDict = {'aVF.jpg': [], 'aVL.jpg': [], 'aVR.jpg': [], 'lead I.jpg': [], 'lead II.jpg': [], 'lead III.jpg': [],
           'V1.jpg': [], 'V2.jpg': [], 'V3.jpg': [], 'V4.jpg': [], 'V5.jpg': [], 'V6.jpg': []}
directoryList = glob.glob('../patient data 2020.3.27/*')
categories = ['PVC RBBB', 'PVC LBBB', 'sinus']

passCnt = 0
flag = True
for dirs in directoryList:
    for _dir in glob.glob(dirs + '/*'):
        patient = _dir.split('/')[-1]
        feature = []
        label = ''
        cnt = 0
        
        for category in categories:
            if category in patient:
                label = category
                cnt+=1
                break
        if cnt == 0:
            break
            
        print('start: ', _dir)
        passCnt+=1
        flag=True
        for lead, frame in ecgDict.items():
            target = glob.glob(_dir + '/' + lead)
            if len(target) != 1:
                flag=False
                print('this is none target', _dir + '/' + lead)
                print('count', passCnt)
                break
        
            feature = dimReduction(cutUseless(target[0]))
            frame.append(feature)
            
        if flag:
            Y.append(label)


        
```
