```python
DIrs = ['../patient data 2020.3.27', '../patient data 2']
cnt = 0

X = []
y = np.array([])
targets = ['sinus', 'sinus lbbb', 'sinus rbbb', 'pvc lbbb', 'pvc rbbb', 'vt lbbb', 'vt rbbb']
target=''

for Dir in DIrs:
    workDIr = os.path.abspath(Dir)
    print(workDIr)
    for (path, dir, files) in os.walk(workDIr):
#         print(path)

#         for dirname in dirnames:
#             print("\t", dirname)

        imgs = []
        for filename in files:
            if filename == 'RpeakV1.png':
                continue
            ext = os.path.splitext(filename)[-1]
            if ext== '.png':
                imgs.append(filename)
        if len(imgs)>0:
            print(path)
            imgs = sorted(imgs, key = lambda e: int(''.join(list(filter(str.isdigit, e)))))
            
            for label in targets:
                if label in path.split('/')[-1].lower():
                    target = label
                    
            feature = np.array([])
            for filename in imgs:
                print(filename)
                if len(feature) == 15:
                    break
                im = cv2.imread(path + '/' + filename, cv2.IMREAD_GRAYSCALE)
                im = cv2.resize(im, dsize=(32, 128),interpolation=cv2.INTER_LINEAR)
                im = im.reshape(128,32,1)
                if len(feature) == 0:
                    feature = np.array([im])
                else:
                    feature = np.vstack((feature, [im]))
                    
            if len(feature)<15:
                dumy = np.zeros(((1,128,32,1)))
                for  i in range(15-len(feature)):
                    feature = np.vstack((feature, dumy))
                    
            if len(X) == 0:
#                 X = np.array([feature])
                X.append(feature)
                y = np.array(target)
            else:
#                 X = np.vstack((X, [feature]))
                X.append(feature)
                y = np.vstack((y, target))
                        
```

```python
from keras import layers, models
from keras.applications import VGG16
from keras import Input
from keras.models import Model
from keras.preprocessing.image import ImageDataGenerator
from keras import optimizers, initializers, regularizers, metrics
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
import os
from glob import glob
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from scipy import io
from sklearn.metrics import f1_score, accuracy_score
def get_model():
    input_tensor = Input(shape=(None, 128, 32, 1), dtype='float32', name='input')
    
    x = layers.Conv3D(64, (3, 3, 3), activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01))(input_tensor)
    x = layers.Conv3D(64, (3, 3, 3), activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.MaxPooling3D((2,2,2))(x)
    
    x = layers.Conv3D(128, (3, 3, 3), activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.Conv3D(128, (3, 3, 3), activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.MaxPooling3D((2,2,2))(x)
    
    x = layers.Conv3D(256, (3, 3, 3), activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.Conv3D(256, (3, 3, 3), activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.Conv3D(256, (3, 3, 3), activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.MaxPooling3D((2,2,2))(x)
    
    x = layers.Conv3D(512, (3, 3, 3), activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.Conv3D(512, (3, 3, 3), activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.Conv3D(512, (3, 3, 3), activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.MaxPooling3D((2,2,2))(x)
    
    x = layers.Conv3D(512, (3, 3, 3), activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.Conv3D(512, (3, 3, 3), activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.Conv3D(512, (3, 3, 3), activation='relu', padding='same', kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01))(x)
    x = layers.MaxPooling3D((2,2,2))(x)
    
    x = layers.Dense(4096, kernel_initializer='he_normal')(x)
    x = layers.Dense(4096, kernel_initializer='he_normal')(x)
    output_tensor = layers.Dense(7, activation='softmax')(x)
    
    myvgg = Model(input_tensor, output_tensor)
    myvgg.compile(loss='sparse_categorical_crossentropy', optimizer=optimizers.RMSprop(lr=2e-5), metrics=['acc'])
    myvgg.summary()
    return myvgg
```
